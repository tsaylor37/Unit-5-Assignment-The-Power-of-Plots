##### Objective: Anaylize data for new opportunities within the ride sharing market. By creating a bubble plot that shows the relationship between: Average Fare (dollars) Per City, Total Number of Rides Per City, Total Number of Drivers Per City, and City Type (Urban, Suburban, Rural). Also, create three pie charts that shows: Percent of Total Fares by City Type, Percent of Total Rides by City Type, Percent of Total Drivers by City Type. All data is being pulled from the city_data.csv and ride_data.csv data sets.

```python
# Magic function used to show graphs.
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load files
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read City and Ride data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Check city data
city_data

```

```python
#Drop duplicate cites from city_data
city_data = city_data.drop_duplicates('city')
#Check city_data for dropped 
city_data
```

```python
#Check ride data
ride_data
```

```python
# Combine the data into single dataset by city
CombinedCityRideData = pd.merge(ride_data, city_data,
                                how='outer', on='city')
# Display the merged table
CombinedCityRideData
```

## Bubble Plot of Ride Sharing Data

```python
# Create separate data frames based on city types - Urban, Suburban and Rural
UrbanCity = CombinedCityRideData.loc[(CombinedCityRideData["type"] == "Urban")]
SuburbanCity = CombinedCityRideData.loc[(CombinedCityRideData["type"] == "Suburban")]
RuralCity = CombinedCityRideData.loc[(CombinedCityRideData["type"] == "Rural")]
```

```python

# Per city calculations of Average Fare, Total Rides and Total Drivers for Urban cities
UrbanAverageFarePerCity = UrbanCity.groupby(["city"]).mean()["fare"]
UrbanTotalRides = UrbanCity.groupby(["city"]).count()["ride_id"]
UrbanTotalDrivers = UrbanCity.groupby(["city"])["driver_count"].value_counts()

# Per city calculations of Average Fare, Total Rides and Total Drivers for Suburban cities
SubAverageFarePerCity = SuburbanCity.groupby(["city"]).mean()["fare"]
SubTotalRidesCity = SuburbanCity.groupby(["city"]).count()["ride_id"]
SubTotalDriversCity = SuburbanCity.groupby(["city"])["driver_count"].value_counts()

# Per city calculations of Average Fare, Total Rides and Total Drivers for Rural cities
RuralAverageFarePerCity = RuralCity.groupby(["city"]).mean()["fare"]
RuralTotalRidesCity = RuralCity.groupby(["city"]).count()["ride_id"]
RuralTotalDriversCity = RuralCity.groupby(["city"])["driver_count"].value_counts()
```

```python
# Create scatter plots for Urban, Suburban and Rural cities
plt.scatter(UrbanTotalRides, UrbanAverageFarePerCity,s=UrbanTotalDrivers*10,
            marker ='o', facecolors ="lightcoral", edgecolors="black",alpha = 0.55, linewidth = 2, label="Urban")

plt.scatter(SubTotalRidesCity, SubAverageFarePerCity,s=SubTotalDriversCity*10,
            marker ='o', facecolors ="lightskyblue", edgecolors="darkblue",alpha = 0.55, linewidth = 2, label="Suburban")

plt.scatter(RuralTotalRidesCity, RuralAverageFarePerCity,s=RuralTotalDriversCity*10,
            marker ='o', facecolors ="gold", edgecolors="black",alpha = 0.55, linewidth = 2, label="Rural")

# Chart title
plt.title("Pyber Ride Sharing Data (2016)")
# x label
plt.xlabel("Total Numbers of Rides (Per City)")
# y label
plt.ylabel("Average Fare($)")
# legend 
plt.legend(loc="upper right",frameon = True, edgecolor = "black",title = "City Types")
#Text label notes regarding circle size
plt.text(42,35,"Note:")
plt.text(42,33,"Circle size correlates with driver count per city.")
#Show grid
plt.grid()
# Save an image of the chart and print to screen
plt.savefig("graphs/Pyber Ride Sharing Data.png")
# Show figure
plt.show()
```

## Total Fares by City Type

```python
# Calculating sum of fares per city type
TypeTotalFares = CombinedCityRideData.groupby(["type"]).sum()["fare"]
TypeTotalFares
```

```python
# Setting total fares per city type to an index
TypeTotalFares.index
```

```python
# Build Pie Chart
colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,0.1)

# Create pie chart based data above
plt.pie(TypeTotalFares, explode=explode, labels=TypeTotalFares.index, colors=colors,autopct="%1.1f%%",shadow=True, 
        startangle=155)
# Create Title
plt.title("% of Total Fares by City Type")
# Create axes which are equal to form a circle
plt.axis("equal")
# Save an image of chart to graphs folder 
plt.savefig("graphs/Total Fares by City Type.png")
# Show graph
plt.show()
```
